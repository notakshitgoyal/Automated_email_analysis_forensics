```python
import mailparser
import requests
import hashlib

# IPinfo API key for IP geolocation (replace with your API key)
IPINFO_API_KEY = ""  # Get from https://ipinfo.io/

# Function to parse email and extract metadata
def parse_email(file_path):
    # Load email file
    mail = mailparser.parse_from_file(file_path)

    # Print basic metadata
    print(f"Subject: {mail.subject}")
    print(f"From: {mail.from_}")
    print(f"To: {mail.to}")
    print(f"Date: {mail.date}")
    print(f"Message ID: {mail.message_id}")
    print(f"Return Path: {mail.return_path}")
    
    # Received headers for IP tracking
    print("\n--- Received Headers ---")
    for received in mail.received:
        print(received)

    # Analyze email body (text or HTML)
    if mail.text_plain:
        print("\n--- Plain Text Body ---")
        print(mail.text_plain)
    if mail.text_html:
        print("\n--- HTML Body ---")
        print(mail.text_html)
    
    # Check for and analyze attachments
    if mail.attachments:
        print("\n--- Attachments Found ---")
        for attachment in mail.attachments:
            print(f"Filename: {attachment['filename']}")
            print(f"Content Type: {attachment['mail_content_type']}")
            
            # Safely check if 'size' key exists
            size = attachment.get('size', 'Unknown')
            print(f"Size: {size} bytes")
            
            hash_attachment(attachment['payload'], attachment['filename'])
    else:
        print("\nNo attachments found.")
    
    # IP Geolocation Analysis
    print("\n--- IP Geolocation ---")
    ip_addresses = extract_ip_addresses(mail.received)
    for ip in ip_addresses:
        get_ip_info(ip)

# Function to extract IP addresses from "Received" headers
def extract_ip_addresses(received_headers):
    ip_list = []
    for header in received_headers:
        if 'from' in header:
            parts = header['from'].split()
            for part in parts:
                if is_valid_ip(part):
                    ip_list.append(part)
    return ip_list

# Function to check if string is a valid IP address
def is_valid_ip(ip):
    parts = ip.split('.')
    return len(parts) == 4 and all(part.isdigit() and 0 <= int(part) <= 255 for part in parts)

# Function to get IP geolocation information using IPinfo API
def get_ip_info(ip_address):
    url = f"https://ipinfo.io/{ip_address}?token={IPINFO_API_KEY}"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        print(f"IP: {data['ip']}, Location: {data.get('city', 'N/A')}, {data.get('region', 'N/A')}, {data.get('country', 'N/A')}, Org: {data.get('org', 'N/A')}")
    else:
        print(f"Could not retrieve information for IP: {ip_address}")

# Function to hash attachment for integrity verification
def hash_attachment(payload, filename):
    # Convert payload to bytes
    file_data = payload.encode('utf-8')
    
    # Compute MD5, SHA-1, and SHA-256 hashes
    md5_hash = hashlib.md5(file_data).hexdigest()
    sha1_hash = hashlib.sha1(file_data).hexdigest()
    sha256_hash = hashlib.sha256(file_data).hexdigest()

    print(f"\n--- Hashing for {filename} ---")
    print(f"MD5: {md5_hash}")
    print(f"SHA-1: {sha1_hash}")
    print(f"SHA-256: {sha256_hash}")

# Function to search for specific keywords in the email body
def keyword_search(mail, keywords):
    content = ' '.join(mail.text_plain + mail.text_html)
    found_keywords = {keyword: content.lower().count(keyword.lower()) for keyword in keywords if keyword.lower() in content.lower()}
    
    print("\n--- Keyword Search Results ---")
    if found_keywords:
        for keyword, count in found_keywords.items():
            print(f"Keyword '{keyword}' found {count} times.")
    else:
        print("No keywords found.")

# Main function to run the analysis
def run_analysis(file_path, keywords):
    parse_email(file_path)
    print("\n--- Keyword Analysis ---")
    mail = mailparser.parse_from_file(file_path)
    keyword_search(mail, keywords)


dummy_email_file = 'test.eml'

# Replace the keyword list with your specific keywords
keywords_to_search = [
    'abducted', 'President‚Äôs daughter', 'secret location', 
    '1 Billion CoCanDs', 'cash', 'spaceship', 'autonomous bots',
    'puzzle', 'solve the puzzle', 'Don‚Äôt Trust Your Eyes', 
    'light minutes', 'waiting for the cash'
]

run_analysis(dummy_email_file, keywords_to_search)
```
